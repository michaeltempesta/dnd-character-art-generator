```python
# ğŸ”§ Step 1: Configure Your Setup
# Choose your options before we start the installation process

import subprocess
import sys
import os
from pathlib import Path
import torch
import gradio as gr

# =============================================================================
# ğŸ¯ CONFIGURATION OPTIONS - Choose what you want:
# =============================================================================

# ğŸ“ Google Drive Integration
USE_GOOGLE_DRIVE = True  # Set to False if you don't want Google Drive integration

# ğŸ”‘ API Keys (Optional but Recommended)
OPENAI_API_KEY = ""  # For AI prompt enhancement - get from https://platform.openai.com/api-keys
HUGGINGFACE_TOKEN = ""  # For model access - get from https://huggingface.co/settings/tokens

# ğŸ¨ Art Generation Preferences
DEFAULT_QUALITY = "high"  # Options: "low", "medium", "high", "ultra"
AUTO_OPTIMIZE = True  # Automatically optimize settings for your GPU

# ğŸ”’ Security Settings
APP_USERNAME = "user"  # Change this for security
APP_PASSWORD = "secure-password"  # Change this for security

# =============================================================================
# ğŸ“‹ Your Current Configuration:
# =============================================================================

print("ğŸ¯ CONFIGURATION SUMMARY:")
print("=" * 50)
print(f"ğŸ“ Google Drive: {'âœ… ENABLED' if USE_GOOGLE_DRIVE else 'âŒ DISABLED'}")
print(f"ğŸ¤– OpenAI AI Enhancement: {'âœ… ENABLED' if OPENAI_API_KEY else 'âŒ DISABLED'}")
print(f"ğŸ”‘ Hugging Face Access: {'âœ… ENABLED' if HUGGINGFACE_TOKEN else 'âŒ DISABLED'}")
print(f"ğŸ¨ Default Quality: {DEFAULT_QUALITY}")
print(f"âš¡ Auto-Optimize: {'âœ… ENABLED' if AUTO_OPTIMIZE else 'âŒ DISABLED'}")
print(f"ğŸ”’ App Security: Username='{APP_USERNAME}', Password='{APP_PASSWORD}'")
print("=" * 50)

if not USE_GOOGLE_DRIVE:
    print("âš ï¸  Google Drive is disabled - files will be saved locally only")
if not OPENAI_API_KEY:
    print("â„¹ï¸  OpenAI key not provided - using default prompts (still works great!)")
if not HUGGINGFACE_TOKEN:
    print("â„¹ï¸  Hugging Face token not provided - using public access (may be slower)")

print("\nğŸš€ Ready to proceed? Run the next cell to start setup!")
```

```python
# ğŸš€ Step 2: Automatic Setup
# This cell handles everything based on your configuration above

def setup_google_drive():
    """Setup Google Drive integration for automatic file saving"""
    if not USE_GOOGLE_DRIVE:
        print("ğŸ“ Google Drive integration disabled by user choice")
        return "/content/outputs"
    
    print("ğŸ“ Setting up Google Drive integration...")
    
    try:
        from google.colab import drive
        drive.mount('/content/drive')
        
        # Create output directories in Google Drive
        drive_output_dir = "/content/drive/MyDrive/DnD_Character_Art"
        os.makedirs(drive_output_dir, exist_ok=True)
        os.makedirs(f"{drive_output_dir}/generated_images", exist_ok=True)
        os.makedirs(f"{drive_output_dir}/masks", exist_ok=True)
        os.makedirs(f"{drive_output_dir}/comfyui_exports", exist_ok=True)
        
        print("âœ… Google Drive mounted successfully!")
        print(f"ğŸ“ Output directory: {drive_output_dir}")
        print("ğŸ¨ All generated art will be automatically saved to your Google Drive")
        
        return drive_output_dir
        
    except Exception as e:
        print(f"âš ï¸  Google Drive setup failed: {e}")
        print("â„¹ï¸  Files will be saved locally in /content/outputs")
        return "/content/outputs"

def setup_environment():
    """One-click setup with intelligent defaults based on user configuration"""
    print("ğŸš€ Setting up your AI art studio...")
    print("=" * 50)
    
    # Setup Google Drive if enabled
    output_dir = setup_google_drive()
    
    # Clone repository if not exists
    if not os.path.exists("dnd-character-art-generator"):
        print("ğŸ“¥ Cloning repository...")
        subprocess.run(["git", "clone", "https://github.com/michaeltempesta/dnd-character-art-generator.git"])
    
    os.chdir("dnd-character-art-generator")
    
    # Install dependencies
    print("ğŸ“¦ Installing dependencies...")
    subprocess.run([sys.executable, "-m", "pip", "install", "-r", "requirements.txt"])
    
    # Fix for CLIPTextModel compatibility issue
    print("ğŸ”§ Installing compatible diffusers version...")
    subprocess.run([sys.executable, "-m", "pip", "install", "diffusers==0.24.0", "transformers==4.35.0", "--force-reinstall"])
    
    # Auto-detect GPU and optimize settings
    gpu_available = torch.cuda.is_available()
    if gpu_available:
        print(f"âœ… GPU detected: {torch.cuda.get_device_name()}")
        print(f"âœ… VRAM: {torch.cuda.get_device_properties(0).total_memory / 1e9:.1f}GB")
    else:
        print("âš ï¸  No GPU detected - using CPU (slower but functional)")
    
    # Initialize with user's configuration
    config = {
        "gpu_available": gpu_available,
        "auto_optimize": AUTO_OPTIMIZE,
        "default_quality": DEFAULT_QUALITY,
        "output_directory": output_dir,
        "google_drive_enabled": output_dir.startswith("/content/drive"),
        "openai_key": OPENAI_API_KEY if OPENAI_API_KEY else None,
        "hf_token": HUGGINGFACE_TOKEN if HUGGINGFACE_TOKEN else None,
        "app_username": APP_USERNAME,
        "app_password": APP_PASSWORD
    }
    
    print("âœ… Setup complete! Your AI art studio is ready.")
    print(f"ğŸ“ All outputs will be saved to: {output_dir}")
    print("=" * 50)
    return config

# Run the setup
config = setup_environment()
```
